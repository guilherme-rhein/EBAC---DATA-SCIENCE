**1. Monte um passo a passo para o Bagging:**

> O **Bagging (Bootstrap Aggregating)** é um procedimento procedimento composto pelos seguintes passos:
> 1. **Bootstrap:** No processo de bootstrap, são geradas amostras aleatórias de treinamento com reposição a partir do conjunto de dados original. Cada amostra bootstrap mantém o mesmo tamanho do conjunto de dados original, mas permite que cada exemplo seja selecionado várias vezes. Esse método visa introduzir variabilidade nos conjuntos de dados utilizados para treinar modelos subsequentes.<br>

> 2. **Modelagem** Para cada amostra Bagging gerada, um modelo de Machine Learning é treinado. Isso implica na criação de vários modelos, onde cada um é treinado em uma amostra com resultados individuais.<br>

> 4. **Aggregating:** Os resultados dos modelos individuais são avaliados para que seja dado uma previsão final com base em cada resultado de cada modelo. Quando estamos lidando com questões de classificação, a previsão final será dada pela moda dos resultados dos modelos. Quando os dados são de regressão, a previsão final será a média dos resultados dos modelos.

<br>**1. Explique com suas palavras o Bagging:**

> O ***Bagging (Bootstrap Aggregating)*** é uma técnica de ensemble que combina várias instâncias de um mesmo modelo de aprendizado de máquina para melhorar a precisão e robustez do modelo final. 
> 1. ***Amostragem Bootstrap:*** Para criar diferentes conjuntos de treinamento, o Bagging utiliza amostragem bootstrap. Isso significa que para cada árvore no RandomForest, são amostrados aleatoriamente exemplos do conjunto de treinamento com substituição, o que significa que um exemplo pode ser escolhido mais de uma vez.<br>

> 2. ***Construção de Árvores:*** Para cada conjunto de treinamento amostrado, uma árvore de decisão é construída. Cada árvore é treinada de forma independente e pode aprender padrões diferentes nos dados.<br>

> 4. ***Agregação:*** Após a construção de todas as árvores, a predição final é feita agregando as previsões de cada árvore. No caso de classificação, isso geralmente envolve uma votação majoritária, enquanto em regressão, pode envolver a média das previsões.
>
>A principal ideia por trás do Bagging é reduzir a variância do modelo, tornando-o mais robusto ao reduzir a sensibilidade a padrões específicos nos dados de treinamento. Isso é particularmente eficaz quando o modelo é propenso a overfitting, melhorando a generalização e desempenho do modelo.

```python
import pandas as pd

from sklearn.datasets import load_iris
from sklearn.datasets import load_diabetes

from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

<br>**Aplicando o processo:**

```python
df = pd.DataFrame(data=load_iris().data, columns=load_iris().feature_names)
df['target'] = load_iris().target
df.head(2)
```

```python
df_train, df_test = train_test_split(df, test_size = 0.20)

X_test = df_test.drop(['target'], axis=1)
y_test = df_test['target'].rename('y_test')

y_train = df_train['target'].rename('y_train')
```

```python
y_pred_bagging = {}
for i in range(10):
    df_train_bagging = df_train.sample(n=len(df_train), replace=True)
    X_train = df_train_bagging.drop(['target'], axis=1)
    
    clf = DecisionTreeClassifier()
    clf.fit(X_train, y_train)

    y_pred_bagging.update({i:clf.predict(X_test)})

predict = pd.DataFrame(y_pred_bagging)
predict = predict.mode(axis=1).rename(columns={0:'y_pred'}).drop(1, axis=1)
predict['y_pred'] = predict['y_pred'].astype(int)
predict['y_test'] = y_test.reset_index(drop=True)
```

```python
print('accuracy score:', accuracy_score(predict.y_test,predict.y_pred))
predict
```
