## Anlise da probabilidade de retorno

Considere somente os dados de janeiro de 2020 em diante para esta anlise.

````python
# Selecionando perodo 
df_slct = df['2020-01-01':]

# Retirando 'Data Emisso' do index
df_slct.dropna(inplace=True)
df_slct.reset_index(inplace=True)
df_slct['Data Emisso'].unique()
````

## Clculo do WOE

- Calcule o WOE das categorias, sendo que o evento em questo  a nota ter retorno>0.
- Analise este WOE por tabela e por grfico.

````python
tab = pd.crosstab(df_slct['categoria'], df_slct['flag_credito'], margins=True, margins_name='total')

tab['odds'] = tab[1]/tab[0]
tab['odds_ratio'] = tab['odds']/tab.loc['total', 'odds']
tab['woe'] = np.log(tab['odds_ratio'])
tab
````

````python
import warnings;   warnings.filterwarnings("ignore")

fig, ax = plt.subplots(figsize=(14, 7))
sns.lineplot(ax=ax, data=tab, x='categoria', y='woe', marker='o')
````

## Clculo do *Information Value*

Calcule o *IV* dessa varivel.

````python
tab = pd.crosstab(df_slct['categoria'], df_slct['flag_credito'], margins=True, margins_name='total')

rotulo_evento = tab.columns[0]
rotulo_nao_evento = tab.columns[1]

tab['pct_evento'] = tab[rotulo_evento]/tab.loc['total', rotulo_evento]
tab['pct_nao_evento'] = tab[rotulo_nao_evento]/tab.loc['total', rotulo_nao_evento]
tab['woe'] = np.log(tab.pct_evento/tab.pct_nao_evento)
tab['iv_parcial'] = (tab.pct_evento - tab.pct_nao_evento)*tab.woe

print(tab.iv_parcial.sum())

tab
````

````python
# 0 a 0,02 - vermelho
# 0,02 a 0,1 - vinho
# 0,1 a 0,3 - amarelo
# 0,3 a 0,5 - azul
# 0,5 ou mais - verde

def highlight_iv(val):
    if 0 <= val <= 0.02:
        return 'background-color: red'
    elif 0.02 < val <= 0.1:
        return 'background-color: #800000'  # Vinho
    elif 0.1 < val <= 0.3:
        return 'background-color: yellow'
    elif 0.3 < val <= 0.5:
        return 'background-color: blue'
    elif val > 0.5:
        return 'background-color: green'
    else:
        return ''

# Aplicando a formatao condicional  coluna 'iv_parcial'
styled_tab = tab.style.applymap(highlight_iv, subset=['iv_parcial'])

# Exibindo a tabela estilizada
styled_tab
````
