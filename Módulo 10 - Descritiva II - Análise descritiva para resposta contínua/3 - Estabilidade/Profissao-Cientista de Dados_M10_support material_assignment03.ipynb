````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('previsao_de_renda.csv')
df
````

1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
plt.figure(figsize=(10, 6))

ax = sns.countplot(x='data_ref', hue='sexo', data=df)

ax.set_xticks(list(range(df['data_ref'].nunique())))
ax.set_xticklabels(df['data_ref'].unique(), rotation=90)
ax.set_yticks(list(range(0, 1000, 100)))
plt.show() 
````

````python
plt.figure(figsize=(12, 6))
ax = sns.countplot(x='data_ref', hue='tipo_renda', data=df)

ax.set_xticks(list(range(df['data_ref'].nunique())))
ax.set_xticklabels(df['data_ref'].unique(), rotation=90)
ax.set_yticks(list(range(0, 1000, 100)))
plt.show() 
````

````python
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='data_ref', hue='educacao', data=df)

ax.set_xticks(list(range(df['data_ref'].nunique())))
ax.set_xticklabels(df['data_ref'].unique(), rotation=90)
ax.set_yticks(list(range(0, 810, 100)))
plt.show() 
````

````python
plt.figure(figsize=(13, 6))
ax = sns.countplot(x='data_ref', hue='estado_civil', data=df)

ax.set_xticks(list(range(df['data_ref'].nunique())))
ax.set_xticklabels(df['data_ref'].unique(), rotation=90)
ax.set_yticks(list(range(0, 1200, 100)))
plt.show() 
````

````python
plt.figure(figsize=(14, 8))
ax = sns.countplot(x='data_ref', hue='tipo_residencia', data=df)

ax.set_xticks(list(range(df['data_ref'].nunique())))
ax.set_xticklabels(df['data_ref'].unique(), rotation=90)
ax.set_yticks(list(range(0, 1000, 50)))
plt.show() 
````

Insights:


- Ao analisar dados ao longo do tempo, percebermos que as características qualitativas permanecem basicamente as mesmas. Isso significa que não há grandes mudanças importantes para investigar nesses aspectos, somente com exceção da variável "educacao" em que após o mês de agosto houve uma queda no grupo "Secundário" e aumento no grupo "Superior Completo". Isso pode ser útil para aplicar uma pesquisa e criar novas suposições que tragam transformações significativas ou como ponto de partida para estudos futuros.


- When analyzing data over time, we realize that the qualitative characteristics remain basically the same. This means that there are no major changes to investigate in these aspects, with the exception of the "educacao" variable, in which after the month of August there was a drop in the "Secundário" group and an increase in the "Superior Completo" group. This can be useful to apply research and create new assumptions that bring significant transformations or as a starting point for future studies.






2) Para cada variável quantitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
mean_qtd_filhos = df.groupby('data_ref')['qtd_filhos'].mean()

mean_qtd_filhos.plot(figsize=[10,6], kind='bar')
plt.xlabel('Data')
plt.ylabel('Média da qtd_filhos')
plt.title('Médias da qtd_filhos ao Longo do Tempo')
plt.show()
````

````python
mean_idade = df.groupby('data_ref')['idade'].mean()

mean_idade.plot(figsize=[10,6], kind='bar')
plt.xlabel('Data')
plt.ylabel('Média da idade')
plt.title('Médias da idade ao Longo do Tempo')
plt.show()
````

````python
mean_tempo_emprego = df.groupby('data_ref')['tempo_emprego'].mean()

mean_tempo_emprego.plot(figsize=[10,6], kind='bar')
plt.xlabel('Data')
plt.ylabel('Média de tempo_emprego')
plt.title('Médias de tempo_emprego ao Longo do Tempo')
plt.show()
````

````python
mean_qt_pessoas_residencia = df.groupby('data_ref')['qt_pessoas_residencia'].mean()

mean_qt_pessoas_residencia.plot(figsize=[10,6], kind='bar')
plt.xlabel('Data')
plt.ylabel('Média de qt_pessoas_residencia')
plt.title('Médias de qt_pessoas_residencia ao Longo do Tempo')
plt.show()
````

````python
mean_renda = df.groupby('data_ref')['renda'].mean()

mean_renda.plot(figsize=[10,6], kind='bar')
plt.xlabel('Data')
plt.ylabel('Média de renda')
plt.title('Médias de renda ao Longo do Tempo')
plt.show()
````

Insights: 
 - Sim, os dados das variáveis quantitativas estão bastante consistentes, com pequenas variações. As maiores variações de valores foram observadas em "qtd_filhos" e "renda". Com base nessa análise, parece viável utilizar essas variáveis em um estudo de modelagem, considerando que as variações não são significativas e provavelmente podem ser acomodadas em um modelo estatístico ou de aprendizado de máquina.

- Yes, the data for the quantitative variables are quite consistent, with small variations. The largest variations in values were observed in 'qtd_filhos' (number of children) and 'renda' (income). Based on this analysis, it seems feasible to use these variables in a modeling study, considering that the variations are not significant and can likely be accommodated in a statistical or machine learning model.
