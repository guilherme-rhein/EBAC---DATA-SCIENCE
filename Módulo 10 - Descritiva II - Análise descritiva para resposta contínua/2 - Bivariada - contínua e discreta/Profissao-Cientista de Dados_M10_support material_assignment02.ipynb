````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('previsao_de_renda.csv')
````

1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
df1 = df.copy()
colunas = ['Unnamed: 0', 'index', 'mau','data_ref', 'sexo', 'tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia']
df1 = df1.drop(columns=colunas)
````

````python
correlation_matrix = df1.corr()
correlation_matrix
````

````python
cmap = sns.diverging_palette(h_neg=200, h_pos=350, as_cmap=True, sep = 1, center = 'light')

sns.clustermap(correlation_matrix, figsize=(8, 8), center = 0, cmap=cmap)
````

````python
sns.regplot(data=df1, x='posse_de_veiculo', y='renda')
plt.title(f'Regplot entre "renda" e "posse_de_veiculo"')
plt.show()
````

````python
sns.regplot(data=df1, x='posse_de_imovel', y='renda')
plt.title(f'Regplot entre "renda" e "posse_de_imovel"')
plt.show()
````

````python
ax = sns.pointplot(x="posse_de_veiculo", y="renda",
                   data=df1, dodge=True, errorbar=('ci', 90))
plt.title(f'Pointplot entre "renda" e "posse_de_veiculo"')
plt.show()
````

````python
ax = sns.pointplot(x="posse_de_imovel", y="renda",
                   data=df1, dodge=True, errorbar=('ci', 90))
plt.title(f'Pointplot entre "renda" e "posse_de_imovel"')
plt.show()
````

- A análise realizada nos dados do arquivo "previsao_de_renda.csv" teve como objetivo avaliar a capacidade de posse de imóvel e posse de veículo como preditores da renda. Concluiu-se que, com base nos dados disponíveis, nenhuma das opções, ou seja, nem a posse de imóvel nem a posse de veículo, são as melhores variáveis para prever a renda. A variável que demonstrou ser a melhor preditora de renda foi o "tempo_emprego". Entre as duas opções consideradas, a "posse de veículo" apresentou um desempenho ligeiramente superior como preditora de renda em comparação com a "posse de imóvel". Portanto, se for necessário escolher entre essas duas variáveis como indicadores da renda, a "posse de veículo" é a opção mais indicada com base nos resultados da análise.

- The analysis conducted on the "previsao_de_renda.csv" data aimed to assess the ability of homeownership and vehicle ownership as predictors of income. It was concluded that, based on the available data, neither of the options, namely homeownership nor vehicle ownership, are the best variables for predicting income.
