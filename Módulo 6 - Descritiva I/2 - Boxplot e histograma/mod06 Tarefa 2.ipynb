Vamos praticar agora análise descritiva de variáveis contínuas na base de dados DataSUS.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sinasc_RO_2019.csv')
df
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

**- Utilizando Pandas para o Boxplot:**

```python
df[['IDADEMAE']].boxplot()
plt.show()
```

**- Utilizando Seaborn para o Boxplot:**

```python
palette = sns.color_palette('Set3')
sns.boxplot(y='IDADEMAE', data=df, palette=palette)
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
sns.boxplot(y='PESO', data=df, color='navy')
plt.title('Distribuição de Peso')
plt.xlabel('Valor')
plt.ylabel('Peso')
plt.show()
```

####  3) Faça um histograma da idade da mãe

**- Utilizando Pandas para o histograma:**

```python
df[['IDADEMAE']].plot.hist(bins = 100)
plt.show()
```

**- Utilizando Seaborn para o histograma:**

```python
sns.displot(x='IDADEMAE', data=df, bins = 100, color='green')
plt.title('Distribuição de Idades')
plt.xlabel('Idade')
plt.ylabel('Valor')
plt.grid(axis='y', linestyle='--', alpha=0.8)
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
sns.displot(x='PESO', data=df, bins=80, color='black')
plt.title('Distribuição de Peso')
plt.xlabel('Peso')
plt.ylabel('Valor')
plt.show()
```
