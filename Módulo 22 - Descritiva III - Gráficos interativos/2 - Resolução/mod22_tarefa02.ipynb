#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio padrão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
3. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

<br>1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu


````python
fech_med = df['Adj Close'].rolling(90).mean()
fech_med = fech_med.stack().reset_index()
fech_med.columns = ['data', 'ativo', 'valor']
fech_med
````

<br>2. Utilizando a fução ```rolling()```, calcule o desvio padrão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima

````python
fech_std = df['Adj Close'].rolling(90).std()
fech_std = fech_std.stack().reset_index()
fech_std.columns = ['data', 'ativo', 'valor']
fech_std
````

<br>3. Monte um gráfico de linhas para as médias móveis dos ativos da sua base

````python
fig = px.line(fech_med, x="data", y="valor", color='ativo', template='plotly_dark')
fig.show()
````

<br>4. Monte um gráfico de linhas para os desvios padrão em janela móvel

````python
fig = px.line(fech_std, x="data", y="valor", color='ativo', template='plotly_dark')
fig.show()
````

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

<br>1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base

````python
slct = df.xs(key='TAEE11.SA', level=1, axis=1).tail(n=60)
slct
````

<br>2. Monte um gráfico de *candlestick* para esses dados.

````python
graph = {
    'x': slct.index,
    'open': slct.Open,
    'close': slct.Close,
    'high': slct.High,
    'low': slct.Low,
    'type': 'candlestick',
    'name': 'TAEE11.SA',
    'showlegend': True
}

layout = Layout(paper_bgcolor='rgba(0,0,0,0)', 
                 plot_bgcolor='rgba(0,0,0,0)', 
                height=500)



grafico = graph_objects.Figure(data=[graph], 
                               layout_title='Ativo TAEE11', 
                               layout=layout)

grafico.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')
grafico.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')

grafico
````
